```python
# mengambil file citra
from google.colab.patches import cv2_imshow
import cv2 as cv
import numpy as np
image1 = cv.imread('images.jpeg')
image2 = cv.imread('download.jpeg')

citra = np.vstack((image1,))
citra1 = np.vstack((image2, citra))

cv2_imshow(citra1)
```

```python
# menampilkan dengan open cv
cv2_imshow(image1)
```

```python
# menampilkan dengan matploplotlib
import matplotlib.pyplot as plt
plt.axis('off')
plt.imshow(image1)
```

```python
# memeriksa ukuran file
print(image1.size)
print(image1.shape)

print(image2.size)
print(image2.shape)
```

```python
# operasi aritmatika pada citra
citra1 = cv.add(image1, image2)
citra2 = cv.subtract(image1, image2)
citra3 = cv.multiply(image1, image2)
citra4 = cv.divide(image1, image2)
```

```python
# operasi bitwise
citra5 = cv.bitwise_not(image1)
citra6 = cv.bitwise_or(image1, image2)
citra7 = cv.bitwise_and(image1, image2)
citra8 = cv.bitwise_xor(image1, image2)
```
