```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt


class Citra:
    'method menampilkan gambar'
    def viewCV(self, image):
        cv2_imshow(image)

    'method memeriksa ukuran dan shape gambar'
    def checkSizeaAndShape(self, image):
        print('size     : ', image.size)
        print('shape    : ', image.shape)
    
    'method operasi aritmatika '
    def opLogikaAritmatika(self, image1, image2):
        

        # pertambahan
        citra1 = cv.add(image1, image2)
        # kurang
        citra2 = cv.subtract(image1, image2)
        # dikai
        citra3 = cv.multiply(image1, image2)
        # dibagi
        citra4 = cv.divide(image1, image2)

        result = np.hstack((citra1, citra2, citra3, citra4))
        cv2_imshow(result)

    'method operasi logika boolean'
    def opLogikaBoolean(self, image1, image2):

        # operasi not
        citra1 = cv.bitwise_not(image1)
        # bitwise and
        citra2  = cv.bitwise_and(image1, image2)
        # bitwise or
        citra3 = cv.bitwise_or(image1, image2)
        # bitwise xor
        citra4 = cv.bitwise_xor(image1, image2)
        
        result = np.hstack((citra1, citra2, citra3, citra4))
        cv2_imshow(result)
        

    'menampilkan banyak gambar dengan matplotlib'
    def tampil_banyak(slef, *image):
      # 1 mewakili baris 8 banyaknya baris dan kolom dan 0 itu adalah kolomnya
        count = 141
        iterasi = 1
        for i in image:
          
          plt.subplot(count)
          count += 1
          plt.imshow(i)
          strtitle= 'citra ' + str(iterasi)
          iterasi+=1
          plt.title(strtitle)

        
    'method geometri penskalan (sacling)'
    def scaling(self, image):
        (height, width) = image.shape[:2]
        result  = cv.resize(image, (int(width / 2), int(height / 2)),interpolation=cv.INTER_CUBIC)
        cv2_imshow(result)

    
    'method geometri pergeseran (translasi)'
    def translate(self, image):
        (height, width) = image.shape[:2]
        M = np.float32([[1, 0, 100], [0, 1, 50]])
        result = cv.warpAffine(image, M, (width, height))
        cv2_imshow(result)
            
    'operasi geometri rotasi'
    def rotate(self, image):
        citra1 =  np.flipud(image)
        citra2 = np.fliplr(image)
        result = np.hstack((citra1, citra2))
        cv2_imshow(result)
        
```

```python
# mengambil file yang sudah di upload
citra1 = cv.imread('images.jpeg')
citra2 = cv. imread('download.jpeg')
# instance object
pcitra = Citra()
```

```python
# melihat ukuran size dan shape citra
pcitra.checkSizeaAndShape(citra1)
```

```python
# method operasi aritmatika
pcitra.opLogikaAritmatika(citra1, citra2)
```

```python
# method operasi logika boolean
pcitra.opLogikaBoolean(citra1, citra2)
```

```python
# menampilkan banyak citra
pcitra.tampil_banyak(citra1, citra2, citra1, citra2)
```

```python
# methond penskalaan (scalling)
pcitra.scaling(citra1)
```

```python
# method geometri pergeseran (translasi)
pcitra.translate(citra2)
```

```python
# method geometri rotasi (rotate)
pcitra.rotate(citra1)
```
